```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
image = cv2.imread('FINAL.tif', cv2.IMREAD_UNCHANGED)

arr = np.array(image)

arr = arr.flatten()

arr = np.array(np.where(arr <=0.1, 0, arr), dtype=np.int16)

L = []
herbaceous = []
deciduous = []
pervious = []

for i in arr:
    if i!=0:
        L.append(i)
    if i==9:
        herbaceous.append(i)
    if i==6:
        deciduous.append(i)
    if i==2:
        pervious.append(i)
        
```

```python
print(f"RATIO HERBACEOUS : {len(herbaceous)/len(L)}")
print(f"RATIO DECIDUOUS : {len(deciduous)/len(L)}")
print(f"RATIO PERVIOUS : {len(pervious)/len(L)}")
```
